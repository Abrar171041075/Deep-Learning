```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

<h3> Image Show

```python
def plt_imshow(title,image):
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title("Cat Image")
    plt.grid(True)
    plt.show()
```

```python
# argumental dictionary for changing key image dynamically

args = {"image" : "../input/single-cat-image/cat.jpg"}
```

```python
cv2.imread("../input/single-cat-image/cat.jpg") #input from directory
```

<h3> Arithmetic operation using cv2

```python
imageAddCustomValue = cv2.add(np.uint8([200]), np.uint8([100]))
imageSubtractCustomValue = cv2.subtract(np.uint8([50]), np.uint8([100]))
```

```python
print("Maximum pixel value in image{}".format(imageAddCustomValue))
print("Minimum pixel value in image{}".format(imageSubtractCustomValue))
```

<h3> Arithmetic operation using numpy

```python
imageAddCustomValue = np.uint8([200]) + np.uint8([100])
imageSubtractCustomValue = np.uint8([50]) - np.uint8([100])
```

```python
print("Maximum pixel value in image{}".format(imageAddCustomValue))
print("Minimum pixel value in image{}".format(imageSubtractCustomValue))
```

```python
img = cv2.imread(args['image'])
img.shape
```

```python
img
```

```python
# creating a custom matrix on the basis of img shape
# np.ones means all the value of the matrix initially will be one

m = np.ones(img.shape,  dtype= np.uint8) * 100 
m
```

```python
?m  # showing doc-string
```

```python
# adding the custom matrix with original img matrix

add = cv2.add(img,m)
plt_imshow("Added Value",add) # calling function
```

```python
# subtract the custom matrix with original img matrix

sub = cv2.subtract(img,m)
plt_imshow("Subtract Value",sub)  # calling function
```

```python
# creating a custom matrix on the basis of img shape

n = np.ones(img.shape,  dtype= np.uint8) * 60
n
```

```python
# adding the custom matrix with original img matrix

add = cv2.add(img,n)
plt_imshow("Added Value",add) # calling function
```

```python
# subtract the custom matrix with original img matrix

sub = cv2.subtract(img,n)
plt_imshow("Subtract Value",sub)  # calling function
```
