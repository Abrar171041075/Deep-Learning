```python
print('Image Translation')
# read the input image
img = cv2.imread("bird-1.jpg")
# convert from BGR to RGB so we can plot using matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# disable x & y axis
#plt.axis('off')
# show the image
print('Original Img')
plt.imshow(img)
plt.show()

# get the image shape
rows, cols, dim = img.shape #open cv te evabe row,col, dimension input
# transformation matrix for translation
M = np.float32([[1, 0, 50],
                [0, 1, 50],
                [0, 0, 1]])
# apply a perspective transformation to the image
translated_img = cv2.warpPerspective(img, M, (cols, rows))
# disable x & y axis
#plt.axis('off')
# show the resulting image
print('Translated Img')
plt.imshow(translated_img)
plt.show()
# save the resulting image to disk
plt.imsave("bird_translated.jpg", translated_img)
```

```python
print('Img Scaling')
# read the input image
img = cv2.imread("bird-1.jpg")
# convert from BGR to RGB so we can plot using matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# disable x & y axis
#plt.axis('off')
# show the image
plt.imshow(img)
plt.show()
# get the image shape
rows, cols, dim = img.shape
#transformation matrix for Scaling
M = np.float32([[1.5, 0  , 0],
            	[0,   2.0, 0],
            	[0,   0,   1]])
# apply a perspective transformation to the image
scaled_img = cv2.warpPerspective(img,M,(cols*2,rows*2))
# disable x & y axis
#plt.axis('off')
# show the resulting image
plt.imshow(scaled_img)
plt.show()
# save the resulting image to disk
plt.imsave("bird_scaled.jpg", scaled_img)
```

```python
print('Img Rotation')
# read the input image
img = cv2.imread("bird-1.jpg")
# convert from BGR to RGB so we can plot using matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# disable x & y axis
#plt.axis('off')
# show the image
plt.imshow(img)
plt.show()
# get the image shape
rows, cols, dim = img.shape
#angle from degree to radian
angle = np.radians(-10)
#transformation matrix for Rotation
M = np.float32([[np.cos(angle), -(np.sin(angle)), 0],
            	[np.sin(angle), np.cos(angle), 0],
            	[0, 0, 1]])
# apply a perspective transformation to the image
rotated_img = cv2.warpPerspective(img, M, (cols,rows))
# disable x & y axis
#plt.axis('off')
# show the resulting image
plt.imshow(rotated_img)
plt.show()
# save the resulting image to disk
plt.imsave("bird_rotated.jpg", rotated_img)
```

```python
import cv2
import pylab as plt
import numpy as np

def varianceCalculate(average, histgram):
    # return variance
    variance = 0
    for i in range(len(histgram)):
        variance += (histgram[i] - average) ** 2

    variance /= len(histgram)   

    return variance

def averageAndpixelSumCalculate(histgram):
    #return the average and the sum of the number of pixels
    average = pixelSum = 0
    for i in range(len(histgram)):
        pixelSum += histgram[i]             #total number of pixels in each portion
        brightnessValue = histgram[i] * i   #sum of luminance values

    average = brightnessValue / len(histgram)  #mean calculation

    return pixelSum, average

def within_betweenCV(pixelSum1, average1, variance1, pixelSum2, average2, variance2):
    #Returns between-class variance & within-class variance

    betweenClassVariance = (pixelSum1 * pixelSum2 * ((average1 - average2) ** 2) ) / ((pixelSum1 + pixelSum2) ** 2)

    withinClassVariance = (pixelSum1 * variance1 + pixelSum2 * variance2) / (pixelSum1 + pixelSum2)

    return betweenClassVariance, withinClassVariance

def calculateAll(blackList, whiteList): 
    b_size, b_average = averageAndpixelSumCalculate(blackList)  #b_size backgroound pixel sum
    w_size, w_average = averageAndpixelSumCalculate(whiteList)

    b_variance = varianceCalculate(b_average, blackList)
    w_variance = varianceCalculate(w_average, whiteList)

    betweenCV, withinCV = within_betweenCV(b_size, b_average, b_variance, w_size, w_average, w_variance)

    totalVariance = betweenCV + withinCV
    separationMetrics = betweenCV / (totalVariance - betweenCV) #Ratio of between-class variance to within-class variance: Calculate the variance

    return separationMetrics

def main():
    image_path = "/content/abrar.jpg"                               
    image = cv2.imread(image_path, 0)                      

    histgram = cv2.calcHist([image], [0], None, [256], [0, 256])

    size = 256
    listSM = [0 for i in range(size)]                       
    for i in range(size):
        if i != 0 and i != size-1:
            blackList = histgram[0: i]                      #Lower brightness list _blackList
            whiteList = histgram[i: size]                   #Larger list of brightness _whiteList
            listSM[i] = calculateAll(blackList, whiteList)  
        elif i == 0 or i == size-1:
            listSM[i] = 0                                   

    #Threshold from Maximum Separation
    maxValue = 0                                            
    for i in range(size):           
        if listSM[i] > maxValue:
            maxValue = listSM[i]
            maxValueIndex = i                               #Store the index with the highest variance
    print("The Threshold value", maxValueIndex)
    

    #Binarize the image on obtained threshold
    output_otsu = np.zeros((len(image), len(image[0])))    #array for otsu
    for i in range(len(image)):
        for j in range(len(image[0])):
            if image[i][j] > maxValueIndex:
                output_otsu[i][j] = 255                     # 255 ifimage index greater than threshold 
            else:
                output_otsu[i][j] = 0                      


    #Creating a binarized image using the value in the middle of the luminance value of the histogram as a threshold
    average_histgram = int(len(histgram) / 2)               #half the length of hist 
    output_average = image.copy()                           #array to use for simple binarization output
    output_average[output_average >= average_histgram] = 255                  
    output_average[output_average < average_histgram] = 0                     




  
    cv2.imwrite("gray.jpg", image)                          
    cv2.imwrite("average.jpg", output_average)              #static threshold

    cv2.imwrite("otsu.jpg", output_otsu)                    #Otsu method

    from google.colab.patches import cv2_imshow
    cv2_imshow( image)
    cv2_imshow(output_average)
    cv2_imshow(  output_otsu)


    print("static threshold value: ", output_average)
    plt.plot(histgram)                                     
    plt.axvline(x=maxValueIndex, color='red', label='otsu') 
    plt.axvline(x=average_histgram, color='green', label='average')
    plt.legend(loc='upper right')                           #show label
    plt.title("histgram of brightness")                    
    plt.xlabel("brightness")                                
    plt.ylabel("frequency")                                 
    plt.xlim([0, 256])                                      #x-axis range setting
    plt.show()

if __name__ == '__main__':
    main()
```
