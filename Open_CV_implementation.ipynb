```python
def ImageShow(imageTitle, image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(imageTitle)
    plt.grid(False)
    plt.tight_layout()
    plt.show()
```

```python
import os
```

```python
def datasetLoaderFromDirectory():
    
    path = "/content/drive/MyDrive/kaggle_dataset/cat-and-dogs/dataset/training_set/cats"
    i= 1
    fig = plt.figure(figsize=(12, 12))
    
  
    for file in os.listdir(path):
        image = cv2.imread(os.path.join(path, file))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.resize(image, (224,224))
        plt.subplot(5, 5, i)
        plt.title(file)
        plt.xticks([])
        plt.yticks([])
        plt.imshow(image)
        i=i+1
    plt.show()
    
```

Translation

```python
movementInformation = np.float32([[1, 0, 100], [0, 1, 200]])
updatedImageAfterMovement = cv2.warpAffine(image, movementInformation, (image.shape[0], image.shape[0]))
ImageShow("updated Image After Movement", updatedImageAfterMovement)
```

```python
movementInformation =np.float32([[1, 0, -200], [0, 1, -105]])
updatedImageAfterMovement = cv2.warpAffine(image, movementInformation, (image.shape[1], image.shape[0]))
ImageShow("updated Image After Movement", updatedImageAfterMovement)
```

```python
import imutils
updatedImageAfterMovement = imutils.translate(image, 0, -100)
ImageShow("updated Image After Movement", updatedImageAfterMovement)
```
